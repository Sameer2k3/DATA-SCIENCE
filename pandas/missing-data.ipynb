```python
import numpy as np
import pandas as pd
```

finding missing data

```python

data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna().sum()
```

removind missing data

```python
df.dropna()  #removes every row which has any missing data
```

```python
df.dropna(thresh=3)  #thres is used to keep rows (or columns) that have at least a minimum number of non-null values.
```

filling the missing data

```python
df
```

```python
values={'A':0,'B':100,'C':300,'D':400}
df.fillna(value=values)
```
